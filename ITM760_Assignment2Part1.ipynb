```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
G_Altgeld = nx.DiGraph()
```

```python
U = ["U1", "U2", "U3", "U4" , "U5", "U6", 
"U7", "U8", "U9"]
G_Altgeld.add_nodes_from(U)
```

```python
G_Altgeld.add_edges_from([('U1','U2') , ('U1', 'U3') , ('U1','U6') , ('U1','U8'),
                          ('U2', 'U1'), ('U2', 'U4'),
                          ('U3','U4'), ('U3','U5'),('U3','U2'),
                          ('U4','U9'), ('U6','U1'), ('U6','U7'), ('U6','U9'),
                          ('U7','U6'), ('U7','U8'),('U7','U2'), ('U8','U4'),('U9', 'U2'),('U9','U5'),('U9','U8')])
```

```python
i = 0
for name in U:
  i = i+1
  print(["%d:%s = %s"% 
         (i,name,str(len(G_Altgeld.out_edges(name))))])
```

```python
nx.draw_circular(G_Altgeld, node_size=1500, node_color='cyan', with_labels = True)
plt.show()
```

```python
nx.pagerank(G_Altgeld, max_iter=100)
```
