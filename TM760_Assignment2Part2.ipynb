```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np
G_Altgeld = nx.DiGraph()
```

```python
Users = ["User1", "User3", "User4", "User5" , "User6", "User7", 
"User8", "User10", "User11"]
G_Altgeld.add_nodes_from(Users)
```

```python
G_Altgeld.add_edges_from([('User1','User3'), ('User1', 'User10'), ('User1','User6'), ('User1','User8'),
                          ('User3','User4'), ('User3','User5'),('User3','User10'),
                          ('User5','User6'), ('User5','User11'),
                          ('User6','User1'),
                          ('User7','User8'), ('User7','User6'),
                          ('User8','User11'), ('User8','User4'),
                          ('User10', 'User1'),
                          ('User11','User3')])
```

```python
i = 0
for name in Users:
  i = i+1
  print(["%d:%s = %s"% 
         (i,name,str(len(G_Altgeld.out_edges(name))))])
```

```python
nx.draw_circular(G_Altgeld,node_color='blue', with_labels = True)
plt.show()
```

```python
nx.pagerank(G_Altgeld, max_iter=100)
```
